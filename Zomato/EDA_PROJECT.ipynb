```python
ratings = final_df.groupby(["Aggregate rating","Rating color", "Rating text"]).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
##Observations
# 1. When Rating is between 4.5 to 4.9 ---> Excellent
# 2 When Rating are between 4.0 to  3.4 ---> Very good
# 3. When Rating are between 3.5 to  3.9 ---> Average
# 4. When Rating are between 2.5 to  2.9 ---> Average
# 5. When Rating are between 2.0 to  2.4 ---> poor
```

```python
ratings.head()
```

```python
sns.barplot(x='Aggregate rating',y="Rating Count", data=ratings)
```

#Observation of Rating colors and rating count

```python
sns.barplot(x='Aggregate rating',y="Rating Count",hue="Rating color", data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y="Rating Count",hue="Rating color", data=ratings, width= 0.4)
```

```python
## Count Plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow'])
```
